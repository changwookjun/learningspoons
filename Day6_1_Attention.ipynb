## Attention 구현

### Attention without batch

```
import numpy as np
import tensorflow as tf

tf.enable_eager_execution()
```

```
# 나 는 야구 좋아 
# 나    [0.1, 0.1, 0.1]
# 는    [0.2, 0.2, 0.2]
# 야구  [0.3, 0.3, 0.3]
# 좋아  [0.5, 0.5, 0.5]

L, H = 4, 3 # Length , Hidden
```

```
emb = np.array([[0.1, 0.1, 0.1],
               [0.2, 0.2, 0.2],
               [0.3, 0.3, 0.3],
               [0.5, 0.5, 0.5]])
print(emb.shape)
```

```
a = np.random.randn(L,1)
print(a)
```

```
ar = a.repeat(H, axis=1)
print(ar)
```

```
t = emb * ar
print(t)
```

```
c = np.sum(t, axis=1)
print(c)
```

### Attention with batch

```
# 나 는 축구 좋아 
# 나    [0.1, 0.1, 0.1]
# 는    [0.2, 0.2, 0.2]
# 야구  [0.3, 0.3, 0.3]
# 좋아  [0.5, 0.5, 0.5]

# 나 는 야구 좋아 
# 나    [0.1, 0.1, 0.1]
# 는    [0.2, 0.2, 0.2]
# 야구  [0.4, 0.4, 0.4]
# 좋아  [0.5, 0.5, 0.5]
B, L, H = 2, 4, 3
```

```
emb = np.array([[[0.1, 0.1, 0.1],
               [0.2, 0.2, 0.2],
               [0.3, 0.3, 0.3],
               [0.5, 0.5, 0.5]],
               [[0.1, 0.1, 0.1],
               [0.2, 0.2, 0.2],
               [0.4, 0.4, 0.4],
               [0.5, 0.5, 0.5]]] )
print(emb.shape)
```

```
a = np.random.randn(B, L, 1)
print(a)
```

```
ar = a.repeat(H, axis=2)
print(ar)
```

```
t = emb * ar
print(t)
```

```
c = np.sum(t, axis=2)
print(c)
```

### Sotfmax

```
logit = tf.nn.softmax(c, axis=1)
print(logit.numpy())
```
